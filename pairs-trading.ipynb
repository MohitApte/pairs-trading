```python
# Calculate the correlation matrix
correlation_matrix = price_data.corr()
print(correlation_matrix)
```

```python
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns
import matplotlib.pyplot as plt

# Perform agglomerative clustering
clustering_model = AgglomerativeClustering(n_clusters=10, metric='euclidean', linkage='ward')
clustering_model.fit(correlation_matrix)

# Add cluster labels to the correlation matrix
clusters = clustering_model.labels_
correlation_matrix['Cluster'] = clusters

# Plot the dendrogram
sns.clustermap(correlation_matrix.iloc[:, :-1], method='ward', cmap='coolwarm', figsize=(10, 10))
plt.show()
```

```python
from sklearn.decomposition import PCA

# Perform PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(correlation_matrix.iloc[:, :-1])

# Create a DataFrame with PCA results
pca_df = pd.DataFrame(data=pca_result, columns=['PCA1', 'PCA2'])
pca_df['Cluster'] = clusters

# Plot the PCA results
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=pca_df, palette='viridis')
plt.title('PCA of Stock Correlations')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# Example: Select pairs from the same cluster
selected_pairs = []
for cluster in set(clusters):
    cluster_stocks = correlation_matrix[correlation_matrix['Cluster'] == cluster].index
    if len(cluster_stocks) > 1:
        for i in range(len(cluster_stocks)):
            for j in range(i + 1, len(cluster_stocks)):
                selected_pairs.append((cluster_stocks[i], cluster_stocks[j]))

print("Selected Pairs:", selected_pairs)
```

```python
import numpy as np
from itertools import combinations

# Function to calculate Euclidean distance
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))

# Extract PCA coordinates
pca_coords = pca_df[['PCA1', 'PCA2']].values

# Calculate distances between all pairs
distances = []
pairs = list(combinations(pca_df.index, 2))

for pair in pairs:
    stock1, stock2 = pair
    point1 = pca_coords[pca_df.index.get_loc(stock1)]
    point2 = pca_coords[pca_df.index.get_loc(stock2)]
    distance = euclidean_distance(point1, point2)
    distances.append((stock1, stock2, distance))

# Sort pairs by distance
sorted_pairs_by_distance = sorted(distances, key=lambda x: x[2])

# Select the best two pairs based on proximity in PCA space
best_pairs_by_distance = sorted_pairs_by_distance[:2]
print("Best pairs based on proximity in PCA space:", best_pairs_by_distance)
```

```python
# Group stocks by cluster
cluster_groups = {}
for cluster in set(clusters):
    cluster_groups[cluster] = pca_df[pca_df['Cluster'] == cluster].index.tolist()

# Find the closest pairs within each cluster
best_pairs_by_cluster = []

for cluster, stocks in cluster_groups.items():
    if len(stocks) > 1:
        # Calculate distances between all pairs within the cluster
        cluster_pairs = list(combinations(stocks, 2))
        cluster_distances = []
        
        for pair in cluster_pairs:
            stock1, stock2 = pair
            point1 = pca_coords[pca_df.index.get_loc(stock1)]
            point2 = pca_coords[pca_df.index.get_loc(stock2)]
            distance = euclidean_distance(point1, point2)
            cluster_distances.append((stock1, stock2, distance))
        
        # Sort pairs by distance and select the closest pair in the cluster
        sorted_cluster_pairs = sorted(cluster_distances, key=lambda x: x[2])
        best_pairs_by_cluster.append(sorted_cluster_pairs[0])

# Sort all best pairs by distance and select the top two
best_pairs_by_cluster_sorted = sorted(best_pairs_by_cluster, key=lambda x: x[2])
best_pairs_from_clusters = best_pairs_by_cluster_sorted[:2]
print("Best pairs based on cluster membership:", best_pairs_from_clusters)
```

```python
# Combine the best pairs from both methods
combined_best_pairs = best_pairs_by_distance[:1] + best_pairs_from_clusters[:1]
print("Selected best pairs combining both criteria:", combined_best_pairs)
```

```python
from sklearn.decomposition import PCA
import pandas as pd

# Perform PCA with more components
pca = PCA(n_components=5)  # Use 5 components as an example
pca_result = pca.fit_transform(correlation_matrix.iloc[:, :-1])

# Create a DataFrame with PCA results
pca_df = pd.DataFrame(data=pca_result, columns=[f'PCA{i+1}' for i in range(5)])
pca_df['Cluster'] = clusters

# Ensure the index of pca_df is the stock symbols
pca_df.index = correlation_matrix.index

# Display the DataFrame to check
print(pca_df.head())
```

```python
from scipy.spatial.distance import mahalanobis
from itertools import combinations
import numpy as np

# Calculate Mahalanobis distance between all pairs
inverse_covariance_matrix = np.linalg.inv(np.cov(pca_result.T))
distances = []
pairs = list(combinations(pca_df.index, 2))

for pair in pairs:
    stock1, stock2 = pair
    point1 = pca_df.loc[stock1, ['PCA1', 'PCA2', 'PCA3', 'PCA4', 'PCA5']].values
    point2 = pca_df.loc[stock2, ['PCA1', 'PCA2', 'PCA3', 'PCA4', 'PCA5']].values
    distance = mahalanobis(point1, point2, inverse_covariance_matrix)
    distances.append((stock1, stock2, distance))

# Sort pairs by distance
sorted_pairs_by_distance = sorted(distances, key=lambda x: x[2])

# Select the best two pairs based on proximity in PCA space
best_pairs_by_distance = sorted_pairs_by_distance[:2]
print("Best pairs based on Mahalanobis distance in PCA space:", best_pairs_by_distance)
```

```python
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint

# Perform cointegration test
def cointegration_test(pair):
    stock1 = price_data[pair[0]]
    stock2 = price_data[pair[1]]
    result = coint(stock1, stock2)
    return result[1]  # Return the p-value

# Calculate cointegration p-values for all pairs
cointegration_results = []
for pair in pairs:
    p_value = cointegration_test(pair)
    cointegration_results.append((pair[0], pair[1], p_value))

# Sort pairs by p-value (lower p-value indicates stronger cointegration)
sorted_pairs_by_cointegration = sorted(cointegration_results, key=lambda x: x[2])

# Select the best two pairs based on cointegration
best_pairs_by_cointegration = sorted_pairs_by_cointegration[:2]
print("Best pairs based on cointegration:", best_pairs_by_cointegration)
```

```python
import numpy as np

# Calculate the spread
price_data['Spread'] = price_data['MDLZ'] - price_data['HSY']

# Calculate the rolling mean and standard deviation of the spread
window = 30  # Rolling window size
price_data['Spread_Mean'] = price_data['Spread'].rolling(window=window).mean()
price_data['Spread_Std'] = price_data['Spread'].rolling(window=window).std()

# Calculate the Z-score of the spread
price_data['Z_Score'] = (price_data['Spread'] - price_data['Spread_Mean']) / price_data['Spread_Std']

# Display the data with the Z-score
print(price_data.tail())
```

```python
# Define entry and exit signals based on Z-score
entry_threshold = 2
exit_threshold = 0.5

price_data['Long_Entry'] = price_data['Z_Score'] < -entry_threshold
price_data['Short_Entry'] = price_data['Z_Score'] > entry_threshold
price_data['Exit'] = (price_data['Z_Score'] < exit_threshold) & (price_data['Z_Score'] > -exit_threshold)

# Display the signals
print(price_data[['Z_Score', 'Long_Entry', 'Short_Entry', 'Exit']].tail())
```

```python
initial_capital = 100000
positions = pd.DataFrame(index=price_data.index).fillna(0)
portfolio = pd.DataFrame(index=price_data.index).fillna(0)

# Buy MDLZ and short HSY when the long entry signal is triggered
positions['MDLZ'] = np.where(price_data['Long_Entry'], 1, 0)
positions['HSY'] = np.where(price_data['Long_Entry'], -1, 0)

# Short MDLZ and buy HSY when the short entry signal is triggered
positions['MDLZ'] = np.where(price_data['Short_Entry'], -1, positions['MDLZ'])
positions['HSY'] = np.where(price_data['Short_Entry'], 1, positions['HSY'])

# Exit positions when the exit signal is triggered
positions['MDLZ'] = np.where(price_data['Exit'], 0, positions['MDLZ'])
positions['HSY'] = np.where(price_data['Exit'], 0, positions['HSY'])

# Calculate the portfolio value
portfolio['Holdings'] = (positions['MDLZ'] * price_data['MDLZ'] + positions['HSY'] * price_data['HSY']).cumsum()
portfolio['Cash'] = initial_capital - (positions['MDLZ'].diff() * price_data['MDLZ'] + positions['HSY'].diff() * price_data['HSY']).cumsum()
portfolio['Total'] = portfolio['Cash'] + portfolio['Holdings']

# Display the final portfolio value
print(portfolio.tail())
```

```python
import alpaca_trade_api as tradeapi
import pandas as pd
import numpy as np

# Initialize the REST API connection
api = tradeapi.REST(api_key_id, api_secret_key, base_url, api_version='v2')

# Define symbols and timeframe
symbols = ['COST', 'STZ']
timeframe = '1Day'
start_date = '2023-01-01'
end_date = '2024-05-01'

# Fetch historical data for each symbol
data = {}
for symbol in symbols:
    bars = api.get_bars(symbol, timeframe, start=start_date, end=end_date).df
    data[symbol] = bars['close']

# Combine data into a single DataFrame
price_data = pd.DataFrame(data)
price_data.dropna(inplace=True)  # Drop rows with missing values

# Calculate the spread
price_data['Spread'] = price_data['COST'] - price_data['STZ']

# Calculate the rolling mean and standard deviation of the spread
window = 30  # Rolling window size
price_data['Spread_Mean'] = price_data['Spread'].rolling(window=window).mean()
price_data['Spread_Std'] = price_data['Spread'].rolling(window=window).std()

# Calculate the Z-score of the spread
price_data['Z_Score'] = (price_data['Spread'] - price_data['Spread_Mean']) / price_data['Spread_Std']

# Define entry and exit signals based on Z-score
entry_threshold = 2
exit_threshold = 0.5

price_data['Long_Entry'] = price_data['Z_Score'] < -entry_threshold
price_data['Short_Entry'] = price_data['Z_Score'] > entry_threshold
price_data['Exit'] = (price_data['Z_Score'] < exit_threshold) & (price_data['Z_Score'] > -exit_threshold)

# Backtest the strategy
initial_capital = 100000
positions = pd.DataFrame(index=price_data.index).fillna(0)
portfolio = pd.DataFrame(index=price_data.index).fillna(0)

# Buy STZ and short COST when the long entry signal is triggered
positions['STZ'] = np.where(price_data['Long_Entry'], 1, 0)
positions['COST'] = np.where(price_data['Long_Entry'], -1, 0)

# Short STZ and buy COST when the short entry signal is triggered
positions['STZ'] = np.where(price_data['Short_Entry'], -1, positions['STZ'])
positions['COST'] = np.where(price_data['Short_Entry'], 1, positions['COST'])

# Exit positions when the exit signal is triggered
positions['STZ'] = np.where(price_data['Exit'], 0, positions['STZ'])
positions['COST'] = np.where(price_data['Exit'], 0, positions['COST'])

# Calculate the portfolio value
portfolio['Holdings'] = (positions['STZ'] * price_data['STZ'] + positions['COST'] * price_data['COST']).cumsum()
portfolio['Cash'] = initial_capital - (positions['STZ'].diff() * price_data['STZ'] + positions['COST'].diff() * price_data['COST']).cumsum()
portfolio['Total'] = portfolio['Cash'] + portfolio['Holdings']

# Calculate daily returns of the portfolio
portfolio['Daily_Return'] = portfolio['Total'].pct_change()

# Calculate cumulative returns
portfolio['Cumulative_Return'] = (1 + portfolio['Daily_Return']).cumprod() - 1

# Total return
total_return = portfolio['Cumulative_Return'].iloc[-1]
print(f"Total Return: {total_return:.2%}")

# Annualized return
num_days = len(portfolio)
trading_days_per_year = 252
annualized_return = (1 + total_return) ** (trading_days_per_year / num_days) - 1
print(f"Annualized Return: {annualized_return:.2%}")

# Calculate Sharpe Ratio
risk_free_rate = 0.01  # Assume a risk-free rate of 1%
excess_return = portfolio['Daily_Return'] - (risk_free_rate / trading_days_per_year)
sharpe_ratio = (excess_return.mean() / excess_return.std()) * np.sqrt(trading_days_per_year)
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")

# Calculate Maximum Drawdown
roll_max = portfolio['Total'].cummax()
daily_drawdown = portfolio['Total'] / roll_max - 1.0
max_drawdown = daily_drawdown.cummin().iloc[-1]
print(f"Maximum Drawdown: {max_drawdown:.2%}")

# Plot the equity curve
plt.figure(figsize=(12, 6))
plt.plot(portfolio['Total'], label='Total Portfolio Value')
plt.title('Equity Curve')
plt.xlabel('Date')
plt.ylabel('Portfolio Value ($)')
plt.legend()
plt.show()

```
